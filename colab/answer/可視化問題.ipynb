## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
url_2 = "https://raw.githubusercontent.com/shinonome-inc/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv?token=GHSAT0AAAAAAB6BKWXB7TDJFUYOU7HW6RWYY7XV4EA"
df_2 = pd.read_csv(url_2)
```

```python
class_0 = df_2[df_2["class"] == 0]
class_1 = df_2[df_2["class"] == 1]
class_2 = df_2[df_2["class"] == 2]
plt.scatter(class_0['length (cm)'], class_0['width (cm)'], color = 'red', label = 'class_0')
plt.scatter(class_1['length (cm)'], class_1['width (cm)'], color = "blue", label = "class_1")
plt.scatter(class_2['length (cm)'], class_2['width (cm)'], color = "green", label = "class_2")
plt.title('クラス０〜２のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.grid()
plt.legend()
```
